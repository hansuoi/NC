```python
import sympy as sym

from sympy.plotting import plot
sym.init_printing(use_unicode = True)
%matplotlib inline

import numpy as np
import sympy
import math
import matplotlib.pyplot as plt


def logistic(m):
    n = np.arange(0, 1001, 1)
    f_i = np.zeros(1001)

    x = sym.Symbol('x')
    f = sym.Function('f')
    f = m * x * (1 - x)

    x_n = 0.1
    for i in range(1001) :
        f_i[i] = f.subs(x, x_n)
        x_n = f_i[i]
    plt.grid()
    plt.title(m)
    plt.scatter(n, f_i)
    plt.show()


m = np.array([3.2, 3.5, 3.73])
for i in range(3):
    logistic(m[i])
```

```python
import sympy as sym

from sympy.plotting import plot
sym.init_printing(use_unicode = True)
%matplotlib inline

import numpy as np
import sympy
import math
import matplotlib.pyplot as plt


def HAttractor(a = 1.4, b = - 0.3) :
    data = np.zeros((2, 10001))

    data[0][0] = 0.1
    data[1][0] = 0.0

    for i in range(10000) :
        data[0][i + 1] = a - b * data[1][i] - data[0][i] ** 2
        data[1][i + 1] = data[0][i]

    plt.grid()
    plt.scatter(data[0], data[1])
    plt.show()


HAttractor()
```

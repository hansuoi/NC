## 畳み込み積分

### 復習
- *h(x) = f＊g(x) = ∫_(-∞, ∞) f(τ) g(x - τ) dτ*
- フーリエ変換後の*H(ω)*は、*H(ω) = F(ω)G(ω)*
- この畳み込み積分の式によりパーセバルの等式が導け、フーリエ展開でパワーが保存され・フーリエ変換でエネルギーが保存されることがわかる。


- コード内では離散的な処理が行われている。
- すなわち、*(a＊v)[n] = Σ_m a[m] v[n - m]*

```python
import numpy as np
```

```python
a = np.array([0, 1, 2, 3, 4, 5])
v = np.array([0.2, 0.8])
```

numpyの畳み込み積分*np.convolve*には、3つのmodeがあるので見ていく。

1. full (デフォルト。modeを指定しなかった場合自動でこれになる。)
    - vを端から端まで動かしきるまで計算する
    - 要素数はN ＋ M - 1
    - 信号の残響を見れる

```python
np.convolve(a, v, mode = 'full')
```

2. same
    - vがaの端まで行ったら終わり
    - 要素数はmax(M, N)

```python
np.convolve(a, v, mode = 'same')
```

3. valid
    - スタートもゴールも、vはaからはみ出さない
    - 要素数は|a - v| + 1

```python
np.convolve(a, v, mode = 'valid')
```

### ノイズの混じったデータの平滑化
#### 説明
- 平滑化…データにおける重要パターンを、ノイズ等、重要性の低いものを除去しながら見つけ出す方法。

今回は移動平均を使います。
- 区間数n / n項移動平均
    - 元データの i番目の要素を中心としたn項の平均を、平滑化したデータの i番目とする

```python
import matplotlib.pyplot as plt
```

```python
# 0 ~ 10, 項数500の等差数列を生成する
x = np.linspace(0, 10, 500)

# 元の信号
y1 = np.sin(x)

# ノイズ
# 平均 = 0, 分散 = 1 の乱数列を生成する
y2 = y1 + np.random.randn(500) * 0.3

# 移動平均を取るためのvを設定。前後5つの値を用いて平均をとる = 重み0.2による畳み込み積分。
v = np.ones(5) / 5

# グラフ描くからsameで
y3 = np.convolve(y2, v, mode = 'same')


# グラフ描画
plt.plot(x, y1, 'r', linewidth = 3)
plt.plot(x, y2, 'b', linewidth = 1)
plt.plot(x, y3, 'y', linewidth = 1.5)
plt.show()
```

```python
import sympy as sym
from sympy.plotting import plot
import numpy as np
import math
import matplotlib.pyplot as plt
sym.init_printing(use_unicode = True)
%matplotlib inline
```

```python
x = sym.Symbol('x')
f = sym.Function('f')
    
f = x ** 3 - 2 * x - 5
df = sym.Derivative(f).doit()
d2f = sym.Derivative(df).doit()

print("{}\n{}\n{}\n".format(f, df, d2f))
```

```python
# 初期値
x_n1 = 50
x_n2 = 50

n = np.arange(0, 14, 1)
f_i1 = np.zeros(14)
f_i2 = np.zeros(14)

delta = 10 ** (-6)

i = 0
# 誤差の対数を格納していく
f_i1[i] = f.subs(x, x_n1)
f_i2[i] = f.subs(x, x_n2)

while i < 10:
    i += 1
    x_n1 = x_n1 - f.subs(x, x_n1) / df.subs(x, x_n1)
    f_i1[i] = f.subs(x, x_n1)
    x_n2 = x_n2 - df.subs(x, x_n2) / d2f.subs(x, x_n2)
    f_i2[i] = f.subs(x, x_n2)


plt.grid()
plt.ylabel("f(x)")
plt.plot(n, f_i1, 'b')
plt.plot(n, f_i2, 'r')
plt.legend(["1D", "2D"])
plt.show()
```

```python
nemui = np.zeros(5)
for i in range(5):
    nemui[i] = math.log(f_i2[i])
    print(nemui[i])
```

```python
plt.grid()
plt.plot(np.arange(0, 5, 1), nemui, 'b')
plt.show()
```
